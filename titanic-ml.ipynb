```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
train = pd.read_csv("train.csv")
test = pd.read_csv("test.csv")
```

```python
def preprocess(df):
    df["Age"].fillna(df['Age'].median(), inplace=True)
    df["Fare"].fillna(df["Fare"].median(), inplace=True)
    df['Sex'] = df['Sex'].map({'male': 0, 'female': 1})
    df = pd.get_dummies(df, columns=['Embarked'])
    df.drop(['PassengerId', 'Name', 'Ticket', 'Cabin'], axis=1, inplace=True)
    return df
```

```python
train = preprocess(train)
X = train.drop('Survived', axis=1)
y = train['Survived']
```

### **RandomForestClassifier**

```python
# Обучение модели
model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X, y)
```

```python
# Предсказание
passenger_ids = test['PassengerId'].copy()
test = preprocess(test)
RForest_pred = model.predict(test)
```

```python
# Сохранение 
output = pd.DataFrame({'PassengerId': passenger_ids, 'Survived': RForest_pred})

output.to_csv('RForest_submission.csv', index=False)
```

### **GradientBoostingClassifier**

```python
model = GradientBoostingClassifier()
model.fit(X, y)
```

```python
# Обучение модели
GBoost_pred = model.predict(test)
```

```python
# Сохранение 
output = pd.DataFrame({'PassengerId': passenger_ids, 'Survived': GBoost_pred})

output.to_csv('GBoost_submission.csv', index=False)
```
