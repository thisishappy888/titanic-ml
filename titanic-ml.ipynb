```python
import pandas as pd
```

```python
train = pd.read_csv("train.csv")
test = pd.read_csv("test.csv")
```

```python
train.head()
```

```python
test.head()
```

```python
def preprocess(df):
    df["Age"] = df["Age"].fillna(df['Age'].median())
    df["Fare"] = df["Fare"].fillna(df["Fare"].median())
    df['Sex'] = df['Sex'].map({'male': 0, 'female': 1})
    df = pd.get_dummies(df, columns=['Embarked'])
    df = df.drop(['PassengerId', 'Name', 'Ticket', 'Cabin'], axis=1)
    return df
```

```python
train = preprocess(train)
X = train.drop('Survived', axis=1)
y = train['Survived']
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
X_train
```

```python
y_train
```

```python
true_labels = pd.read_csv("gender_submission.csv")
true_labels
```

```python
true_survived = true_labels["Survived"]
true_survived
```

```python
test_processed = preprocess(test)
```

#### **RandomForestClassifier**

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model1 = RandomForestClassifier()
model1.fit(X_train, y_train)
```

```python
predict_RF = model1.predict(test_processed)
```

#### **GradientBoostingClassifier**

```python
from sklearn.ensemble import GradientBoostingClassifier
```

```python
model2 = GradientBoostingClassifier()
model2.fit(X_train, y_train)
```

```python
predict_GB = model2.predict(test_processed)
```

#### **LogisticRegression**

```python
from sklearn.linear_model import LogisticRegression
```

```python
model3 = LogisticRegression()
model3.fit(X_train, y_train)
```

```python
predict_LR = model3.predict(test_processed)
```

#### **Метрики**

```python
from sklearn.metrics import (accuracy_score, precision_score,
                             recall_score, f1_score, confusion_matrix, 
                             classification_report
                             )
```

```python
def calc_metrics(true, pred, model_name):
    print(f"Model: {model_name}")
    print(f"Accuracy: {accuracy_score(true, pred):.2f}")
    print(f"Precision: {precision_score(true, pred):.2}")
    print(f"Recall: {recall_score(true, pred):.2}")
    print(f"F1 Score: {f1_score(true, pred):.2}")
    print(f"\nConfusion Matrix:")
    print(confusion_matrix(true, pred))
    print("\nClassification Report:")
    print(classification_report(true, pred))
```

```python
calc_metrics(true_survived, predict_GB, "GradientBoosting")
```

```python
calc_metrics(true_survived, predict_LR, "LogisticRegression")
```

```python
calc_metrics(true_survived, predict_RF, "RandomForest")
```
